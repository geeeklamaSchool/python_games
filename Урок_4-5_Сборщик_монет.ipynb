````python
import arcade

SCREEN_WIDTH = 725
SCREEN_HEIGHT = 607

class MyGame(arcade.Window):
    """ Наш специальный класс Window """

    def __init__(self):
        """ Инициализатор """
        # Вызываем инициализатор родительского класса
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, "Битва на острове")

        # Отключаем видимость курсора
        self.set_mouse_visible(False)

        # Создаем текстуру заднего фона
        self.background = arcade.load_texture("./Images/water.jpg")
        
        # Переменные для списков спрайтов
        self.player_list = arcade.SpriteList()
        self.coin_list = arcade.SpriteList()
        self.enemies_list = arcade.SpriteList()
        
        # Счетчик монет
        self.counter = 0
        
    def setup(self):
        """ Настраиваем игру и инициализируем переменные. """

        # Задаем спрайт игрока
        self.player_sprite = arcade.Sprite(":resources:images/enemies/fishGreen.png", 0.5)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list.append(self.player_sprite)
        
    def on_draw(self):
        """ Этот метод используется для отрисовки """
        arcade.start_render()
        arcade.draw_lrwh_rectangle_textured(0, 0,SCREEN_WIDTH, SCREEN_HEIGHT,self.background)
   
        # Отрисовываем спрайты
        self.player_list.draw()


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()


main()
````

Чтобы получить случайную позицию монет, мы используем функцию random.randrange(range_value) из библиотеки _random_. <br>

````python
import arcade
import random

SCREEN_WIDTH = 725
SCREEN_HEIGHT = 607
number_coins = 50

class MyGame(arcade.Window):
    """ Наш специальный класс Window """

    def __init__(self):
        """ Инициализатор """
        # Вызываем инициализатор родительского класса
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, "Битва на острове")

        # Отключаем видимость курсора
        self.set_mouse_visible(False)

        # Создаем текстуру заднего фона
        self.background = arcade.load_texture("./Images/water.jpg")

        # Переменные для списков спрайтов
        self.player_list = arcade.SpriteList()
        self.coin_list = arcade.SpriteList()
        self.enemies_list = arcade.SpriteList()

        # Счетчик монет
        self.counter = 0
        
    def setup(self):
        """ Настраиваем игру и инициализируем переменные. """

        # Задаем спрайт игрока
        self.player_sprite = arcade.Sprite(":resources:images/enemies/fishGreen.png", 0.5)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list.append(self.player_sprite)
        
        # Задаем спрайты монет
        for i in range(number_coins):
            coin = arcade.Sprite(":resources:images/items/coinGold.png", 0.3)
            coin.center_x = random.randrange(SCREEN_WIDTH)
            coin.center_y = random.randrange(SCREEN_HEIGHT)
            self.coin_list.append(coin)
        
    def on_draw(self):
        """ Этот метод используется для отрисовки """
        arcade.start_render()
        arcade.draw_lrwh_rectangle_textured(0, 0,SCREEN_WIDTH, SCREEN_HEIGHT,self.background)

        # Отрисовываем спрайты
        self.player_list.draw()
        self.coin_list.draw()


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()


main()
````

Однако класс _Sprites_ имеет некоторые особенности сборки. Когда мы создаем спрайт, мы можем установить скорость:

Чтобы оживить наши спрайты, мы должны вызвать метод _update_ в классе _Sprites_ в методе _update_ класса _MyGame_. Вместо обновления каждого спрайта мы можем обновить весь список - `self.coin_list.update()`



Посмотрите на приведенный ниже код и посмотрите, как выглядит код, который заставляет монеты двигаться.

````python

import arcade
import random

SCREEN_WIDTH = 725
SCREEN_HEIGHT = 607
number_coins = 50
number_enemies = 30

class MyGame(arcade.Window):
    """ Наш специальный класс Window """

    def __init__(self):
        """ Инициализатор """
        # Вызываем инициализатор родительского класса
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, "Битва на острове")

        # Отключаем видимость курсора
        self.set_mouse_visible(False)

        # Создаем текстуру заднего фона
        self.background = arcade.load_texture("./Images/water.jpg")

        # Переменные для списков спрайтов
        self.player_list = arcade.SpriteList()
        self.coin_list = arcade.SpriteList()
        self.enemies_list = arcade.SpriteList()

        # Счетчик монет
        self.counter = 0
        
    def gen_random(self, number, scale, image_path, alist, direction):
        """ Метод, который создает спрайты в случайных местах и добавляет их в список """
        for i in range(number):
            sprite_name = arcade.Sprite(image_path, scale)
            sprite_name.center_x = random.randrange(SCREEN_WIDTH)
            sprite_name.center_y = random.randrange(SCREEN_HEIGHT)
            sprite_name.velocity = (0, direction*random.randrange(1,5))
            alist.append(sprite_name)
            
            
    def setup(self):
        """ Настраиваем игру и инициализируем переменные. """

        # Задаем спрайт игрока
        self.player_sprite = arcade.Sprite(":resources:images/enemies/fishGreen.png", 0.5)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list.append(self.player_sprite)

        # Задаем спрайты монет
        self.gen_random(number_coins, 0.3, ":resources:images/items/coinGold.png", self.coin_list)

        # Задаем спрайты врагов
        self.gen_random(number_enemies, 0.4, ":resources:images/enemies/slimeBlue.png", self.enemies_list)

        
    def on_draw(self):
        """ Этот метод используется для отрисовки """
        arcade.start_render()
        arcade.draw_lrwh_rectangle_textured(0, 0,SCREEN_WIDTH, SCREEN_HEIGHT,self.background)

        # Отрисовываем спрайты
        self.player_list.draw()
        self.coin_list.draw()
        self.enemies_list.draw()
        
    def update(self, delta_time):
        """ Анимация и логика """

        # Вызов обновления всех спрайтов
        self.coin_list.update()
        
        for i in self.coin_list:
            if i.center_y<0:
                i.center_y=SCREEN_HEIGHT
    
    def on_mouse_motion(self, x, y, dx, dy):
        """ Обработка движения мыши """
        # Перемещаем центр спрайта игрока в соответствии с положением мыши по осям x и y.
        self.player_sprite.center_x = x
        self.player_sprite.center_y = y


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()

main()
````

# __<span style = "color: Maroon"> Столкновения </span>__



Методы столкновений в классе _Sprite_:
- `Sprite.collides_with_point((x, y))` - Если граница спрайта достигает точки x, y этот метод возвращает True, в противном случае возвращает False.
- `Sprite.collides_with_sprite(Sprite)` - Если один спрайт сталкивается с другим спрайтом (указанным как параметр в скобках), этот метод возвращает True, в противном случае возвращает False.
- `Sprite.collides_with_list(SpriteList)` - Возвращает список, содержащий все спрайты в SpriteList, которые перекрываются с текущим спрайтом. Если нет перекрывающихся спрайтов, то список будет пустым, то есть его длина будет равна нулю.



Метод столкновения помещен в метод _update_ нашего игрового класса.

Для монет у нас следующая логика:

```python 
coins_hit_list = self.player_sprite.collides_with_list(self.coin_list)

for i in coins_hit_list:
    i.remove_from_sprite_lists() #метод, который удаляет спрайт из списка
    self.counter += 1
```



Для врагов у нас есть следующая логика:

```python 
enemies_hit_list = self.player_sprite.collides_with_list(self.enemies_list)

for i in enemies_hit_list:
    self.counter = 0
```

````python
import arcade
import random

SCREEN_WIDTH = 725
SCREEN_HEIGHT = 607
number_coins = 50
number_enemies = 30

class MyGame(arcade.Window):
    """ Наш специальный класс Window """

    def __init__(self):
        """ Инициализатор """
        # Вызываем инициализатор родительского класса
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, "Битва на острове")

        # Отключаем видимость курсора
        self.set_mouse_visible(False)

        # Создаем текстуру заднего фона
        self.background = arcade.load_texture("./Images/water.jpg")

        # Переменные для списков спрайтов
        self.player_list = arcade.SpriteList()
        self.coin_list = arcade.SpriteList()
        self.enemies_list = arcade.SpriteList()

        # Счетчик монет
        self.counter = 0
        
    def gen_random(self, number, scale, image_path, alist, direction):
        """ Метод, который создает спрайты в случайных местах и добавляет их в список """
        for i in range(number):
            sprite_name = arcade.Sprite(image_path, scale)
            sprite_name.center_x = random.randrange(SCREEN_WIDTH)
            sprite_name.center_y = random.randrange(SCREEN_HEIGHT)
            sprite_name.velocity = (0, direction*random.randrange(1,5))
            alist.append(sprite_name)
            
            
    def setup(self):
        """ Настраиваем игру и инициализируем переменные. """

        # Задаем спрайт игрока
        self.player_sprite = arcade.Sprite(":resources:images/enemies/fishGreen.png", 0.5)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list.append(self.player_sprite)

        # Задаем спрайты монет
        self.gen_random(number_coins, 0.3, ":resources:images/items/coinGold.png", self.coin_list)

        # Задаем спрайты врагов
        self.gen_random(number_enemies, 0.4, ":resources:images/enemies/slimeBlue.png", self.enemies_list, 1)

        
    def on_draw(self):
        """ Этот метод используется для отрисовки """
        arcade.start_render()
        arcade.draw_lrwh_rectangle_textured(0, 0,SCREEN_WIDTH, SCREEN_HEIGHT,self.background)

        # Отрисовываем спрайты
        self.player_list.draw()
        self.coin_list.draw()
        self.enemies_list.draw()
        
        
        
    def update(self, delta_time):
        """ Анимация и логика """

        # Вызов обновления всех спрайтов
        self.coin_list.update()
        self.enemies_list.update()
        
        for i in self.coin_list:
            if i.center_y<0:
                i.center_y=SCREEN_HEIGHT
                
        for i in self.enemies_list:
            if i.center_y>SCREEN_HEIGHT:
                i.center_y=0
                
        coins_hit_list = self.player_sprite.collides_with_list(self.coin_list)

        for i in coins_hit_list:
            i.remove_from_sprite_lists() # метод, который удаляет прайт из списка
            self.counter += 1
            
        enemies_hit_list = self.player_sprite.collides_with_list(self.enemies_list)

        for i in enemies_hit_list:
            self.counter = 0
            
    def on_mouse_motion(self, x, y, dx, dy):
        """ Обработка движения мыши """
        # Перемещаем центр спрайта игрока в соответствии с положением мыши по осям x и y.
        self.player_sprite.center_x = x
        self.player_sprite.center_y = y


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()

main()
````
