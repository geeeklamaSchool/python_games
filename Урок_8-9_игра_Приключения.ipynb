1. Игрок может перемещаться из одной комнаты в другую за ход.
2. Даже если двери нет и игрок идет прямо, он перемещается на расстояние одной комнаты.

Игрок начинает с точки, указанной на карте. Игрок ничего не знает о доме. Единственная информация, которую получает игрок, - это «приветственное сообщение», показанное ниже: <br>

_Вы гуляли по острову и нашли дом. Вы вошли в главные двери, а они закрылись прямо за вашей спиной. Вы заперты! Вы должны найти другой способ выбраться из дома. Не потеряйтесь. Удачи!_



В каждой комнате пользователю предоставляется возможность выбора, куда он может пойти. Обычно есть четыре варианта (север, восток, запад, юг), но не все комнаты открыты  во всех четырех направлениях (см. Карту игры). Игрок начинает в коридоре: <br>

_Вы в коридоре. Есть дверь на восток и на запад от вас. Также можно пройти по коридору на север._

На основании пользовательского ввода игрок перемещается в следующую комнату. Изучите код



````python
def hallway_1():
    print("Вы в коридоре. Есть дверь на восток и на запад от вас. "
          "Также можно пройти по коридору на север.")
    a = input("Куда вы пойдете?")
    while a == "юг":
        print("Вы не пожете пойти на юг")
    if a == "север":
        hallway_2()
    elif a == "запад":
        room_1()
    elif a == "восток":
        room_2()


def welcome():
    print(
        "Вы гуляли по острову и нашли дом. Вы вошли в главные двери, а они закрылись прямо за вашей спиной. "
        "Вы заперты! Вы должны найти другой способ выбраться из дома. Не потеряйся. Удачи!")


def main():
    welcome()
    hallway_1()


main()
````

# __<span style = "color: Maroon"> Библиотека Arcade: ревизия </span>__



Пришло время сделать игру графической. Мы начнем с кода, приведенного ниже. <br>
Посмотрите на приведенный ниже код, чтобы понять, что делает каждый метод.

````python
import arcade
import random

SPRITE_SCALING = 0.5
SCREEN_WIDTH = 450
SCREEN_HEIGHT = 500
SCREEN_TITLE = "Игра Приключения"
MOVEMENT_SPEED = 5


class MyGame(arcade.Window):
    """ Основной класс приложения """

    def __init__(self, width, height, title):
        """ Инициализатор """
        super().__init__(width, height, title)

        # Переменные для комнат
        self.current_room = 0
        self.rooms = None

        # Переменные для игрока
        self.player_sprite = None
        self.player_list = None

        # Другие переменные
        self.win = False
        self.counter = 0

    def setup(self):
        """ Настраиваем игру и инициализируем переменные """
        # Настроиваем игрока
        self.player_sprite = arcade.Sprite(":resources:images/animated_characters/female_person/femalePerson_idle.png",
                                           SPRITE_SCALING)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list = arcade.SpriteList()
        self.player_list.append(self.player_sprite)

        # Наш список комнат
        self.rooms = []

    def on_draw(self):
        """Метод отрисовки"""
        arcade.start_render()
        self.player_list.draw()

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажата клавиша """
        if key == arcade.key.UP:
            self.player_sprite.change_y = MOVEMENT_SPEED
        elif key == arcade.key.DOWN:
            self.player_sprite.change_y = -MOVEMENT_SPEED
        elif key == arcade.key.LEFT:
            self.player_sprite.change_x = -MOVEMENT_SPEED
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = MOVEMENT_SPEED

    def on_key_release(self, key, modifiers):
        """Вызывается, когда пользователь отпускает клавишу """
        if key == arcade.key.UP or key == arcade.key.DOWN:
            self.player_sprite.change_y = 0
        elif key == arcade.key.LEFT or key == arcade.key.RIGHT:
            self.player_sprite.change_x = 0

    def keep_walls(self):
        """Этот метод не дает игроку выйти за пределы экрана"""
        if self.player_sprite.center_x < 0 or self.player_sprite.center_x > SCREEN_WIDTH:
            self.player_sprite.change_x = 0
        elif self.player_sprite.center_y < 0 or self.player_sprite.center_y > SCREEN_HEIGHT:
            self.player_sprite.change_y = 0

    def update(self, delta_time):
        """ Движения и игровая логика """
        self.player_list.update()
        self.keep_walls()


def main():
    """ Метод main """
    window = MyGame(SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_TITLE)
    window.setup()
    arcade.run()


main()
````
