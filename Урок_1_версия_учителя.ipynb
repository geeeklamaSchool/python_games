### <center> <span style="color:Maroon">Типы данных
    
**<span style="color:Teal"> А теперь освежим в памяти несколько типов данных, которые пригодятся в дальнейших уроках.</span>**

**<span style="color:Teal"> В Python нет необходимости объявлять тип переменной вручную (как, например, в C++). Объявление происходит автоматически (это называется динамической типизацией), когда вы присваиваете значение переменной.</span>**

Числовые типы данных:
   - *`integer`* - целое число (не дробная часть), которое может быть положительным (10), отрицательным (-7) или нулем (0).
   - *`float`* - числа с дробными значениями, например 0,5, -4,6.

Типы данных последовательности:
   - *`string`* - последовательность символов, используемая для представления текста. Это может быть одно слово, предложение или целая книга. Чтобы присвоить строку, символы должны быть заключены в кавычки.
   - *`list`* - упорядоченный и изменяемый набор элементов произвольных типов. В Python списки записываются в квадратные скобки, а элементы разделяются запятыми.

  
**<span style="color:teal "> Числовой тип данных Python предназначен для хранения числовых значений. Это неизменяемый тип данных, что означает, что изменение значения числового типа данных приведет к созданию нового объекта в памяти (и удалению старого).</span>**

**<span style="color:Teal"> Следующие типы - это последовательности: строка и список. Строки могут быть представлены в виде списков символов, где каждый символ является элементом. Однако строка является неизменяемым типом данных, а список - изменяемым.</span>**

### <center> <span style="color:Maroon">  Операторы </span>

**<span style="color:teal "> Теперь мы обсудим, что мы можем делать с различными переменными, начиная с чисел. Числа в Python ничем не отличаются от обычных чисел. Они поддерживают набор наиболее распространенных математических операций. Мы также можем сравнивать значения.</span>**

Числа в Python ничем не отличаются от обычных чисел.
    
Какие есть операции с числами:

  - Арифметика (`+`, `-`, `/`, `*`, `**`  возведение в степень, `//` целочисленное диление, `%` модуль, `abs(num)` - абсолютное значение)
  - Сравнение (`>`/`<` меньше/больше, `<=`/`>=` меньше/больше или равно, `==` равно, `!=`неравно, *`x is y`* -  один и тот же обьект ...) - вернет результат *True* или *False*


**<span style="color:teal "> Попрактикуемся с числовыми переменными. Обратите внимание, что числа с плавающей запятой не являются десятичными дробями и используют двоичную компьютерную арифметику, поэтому многие, даже самые простые выражения могут быть вычислены с незначительными ошибками. Однако из-за этих ошибок вполне очевидные операции сравнения не работают должным образом.</span>**

Теперь мы назначим некоторые числовые переменные и будем с ними работать.

Обратите внимание, что суммирование двух первых переменных неверно из-за двоичной компьютерной арифметики. Затем с помощью сравнения укажите на эту ошибку: должно быть *False*, потому что эти значения должны быть равны. </span>**


````python
# зададим значение нескольким переменным
num1 = .78
num2 = 3.

# базовая математика 
a = num1 + num2 # сумма
b = -(num2) # изменение знака
c = (a*b - a**b) % 5 # длинное математическое выражение
print('a =',a,'b =',b,'c =',c) # напечатать результаты
````

**<span style="color:Teal"> Следующие типы - строка и список. Поговорим о том, что мы можем с ними сделать.</span>**
  
**<span style="color:teal "> Оба они представляют собой последовательности. Таким образом, у них обоих есть индексы - у каждого элемента свое место. Нумерация начинается с нуля. Это означает, что у них есть срезы - своего рода диапазон между индексами.</span>**

Строка и список - это последовательности. Таким образом, у них обоих есть индексы и их можно разбить на промежутки.
    

Как использовать индексы для разбития на промежутки:<br>
 - __Indexes__:
    
   - `list1[0]` - элемент с индексом 0 - первым в списке будет 1
   - `list1[-2]` - элемент с индексом -2 - значение будет 4.5

<img src="./Images/Indexes.jpg" width="40%">


 - __Деление на промежутки__:

   - `list1[2:5]` - элементы с третьего по пятый: [3, 'дерево', 4.5]
   - `list1[:2]` -элементы перед вторым (включая): [1, 4]
   - `list1[-2:]` - также допускаются отрицательные индексы - начинаем расчет с конца - два с конца: [4.5, 6]
   - `list1[-6:-3]` - с первого по третий: [1, 4, 3]

<img src="./Images/slices.jpg" width="40%">

**<span style="color:teal "> Учителю: сделаем новую фразу «Счастливые жители острова Питон» - строкой из существующей строки и списка, используя индексы и срезы. Обратите внимание на то, что мы добавляем пробелы.</span>**

**Методы для строк**:
<br>

- `len(string)` - количество всех элементов в строке

<br>  <br> 

**Методы для списков**:

**<span style = "color: teal"> Учителю: объясните ученикам, что если *if* - оператор *True*, то выполняется блок под ним. Когда он равен *False*, мы проверяем оператор *elif* -, если он равен *True*, то выполняется блок *elif*. Если все операторы *False*, программа выполняет блок *else*. </span>**

**<span style = "color: teal"> Обратите внимание на комментарии. Они помогут вам понять логику. </span>**

Условный оператор *`if-elif-else`* - это основной инструмент в Python. Синтаксис приведен ниже, а логика объяснена в комментариях.
    
- Как использовать условный оператор:
```python
if test1: # условное выражение
    state1 # часть, которая выполняется, если оператор if имеет значение True
elif test2: # необязательная часть еще одного условного выражения, если первый оператор if имеет значение False
    state2 # часть, которая выполняется, если elif выражение истинно
else: # необязательная часть, которая выполняется, если оператор if и оператор elif имеют значение False
    state3
``` 
Например:

```python
a = int(input()) # входные данные от пользователя преобразованы в целое число
if a < -5: # проверка if-выражения, что <-5
    print('Низкий') # если это правда, программа печатает "Низкий"
elif -5 <= a <= 5: # если первый if-оператор ложен, мы проверяем еще один оператор: -5 <= a <= 5
    print('Средний') # если второй оператор истинен, программа печатает 'Средний'
else: # если оба оператора if ложны
    print('Высокий') # затем программа напечатает 'High'
``` 

**<span style="color:teal "> Обратите внимание, что *elif* и *else* необязательны. Есть много условий, которые можно написать без них.</span>**

Перейдите к приведенному ниже примеру. В качестве дополнительного условия мы проверим, делится ли введенное число на 3 или 2.

**<span style = "color: teal"> Давайте посмотрим на пример, где мы проверим, делится ли введенное число на 3 или 2 в качестве дополнительных условий. </span>**
    
**<span style = "color: teal"> Учителю: распечатайте код вместе с учащимися. Поскольку мы будем проверять входной номер, мы должны ввести разные параметры. Таким образом, введите «9», «4» и «11», чтобы получить разные результаты. Студенты могут копировать и вставлять. </span>**

````python
# условный оператор:проверка на деление на 3 или на 2
a = int(input()) # входные данные от пользователя преобразованы в целое число
if a % 3 == 0: # проверка if выражение, что a делится на 3
    print('число делится на 3') # если это истинно, программа печатает это сообщение
elif a % 2 == 0: # если первый if-оператор ложен, мы проверяем еще один оператор: что a делится на 2
    print('число делится на 2') # если второй оператор истинен, программа печатает «число делится на 2»
else: # если все операторы if / elif ложны
    print('число не делится на 3 или 2') # затем программа печатает "число не делится на 3 или 2"
#
````

### <center> <span style="color:Maroon"> _Цикл for_ </span>


**<span style = "color: teal"> Цикл *for* выполняет указанный набор инструкций определенное количество раз, которое определяется количеством элементов в наборе. Этот набор может быть итеративным объектом в виде последовательностей - строки или списка. Или мы можем использовать *range()* для выполнения некоторого кода определенное количество раз. </span>**

**<span style = "color: teal"> Учащиеся должны быть знакомы с циклами, поэтому быстро вспомните тему. </span>**

Цикл *for* очень часто используется в Python. Он выполняет указанный набор инструкций указанное количество раз или по указанному списку значений.

`range(start, stop, step)` - встроенная функция, которая возвращает последовательность чисел, начиная с *start* (0 по умолчанию), и увеличивается на *step* (1 по умолчанию) и останавливается перед указанным числом *stop*.

- Как писать *for цикл* используя *range()*:
```python
x = range(6) # составить последовательность чисел от 0 до 5 на 1, используя диапазон
for n in x: # n пройдёт через каждый элемент числа этой последовательности
    print(n) # в результате мы получим распечатанный столбец элементов этой последовательности
``` 

- Как написать *for цикл* для  *списков* и *строк*:
```python
x = 'hello' # строка(последовальность символов)
for n in x: # n пройдет через каждый элемент в x
    print(n) # в результате мы получим распечатанный столбец элементов этой последовательности
``` 

Посмотрим пример использования *for цикла* для списка.

**<span style="color:teal "> Учителю: мы запустим небольшой пример использования *цикла for* для списка. Студенты должны повторять код за вами. Блок цикла будет состоять из одной команды *print*. Этот код печатает каждый элемент списка в столбце.</span>**

````python
# использование цикла с диапазоном
for i in range(5):
    print(i)

# использование цикла for для списка
lst = [ 'На',  'острове', 'Python', 'живут', 'прекрасные', 'жители']
for i in lst:
    print(i) # каждое слово-элемент списка будет печататься с новой строки
    
````

### <center> <span style="color:Maroon"> _Цикл While_ </span>
    
**<span style = "color: Teal"> Цикл *while* позволяет выполнять ту же последовательность действий, пока проверяемое условие истинно. </span>**

**<span style = "color: teal"> Условие записывается перед блоком цикла и проверяется перед выполнением блока цикла. Обычно цикл *while* используется, когда невозможно заранее определить точное количество итераций. </span>**

**<span style = "color: teal"> Синтаксис *цикла while* следующий: </span>**
    
Цикл *while* позволяет выполнять ту же последовательность действий, пока проверяемое условие истинно.

Синтаксис *цикла while*:
```python
k = 0 # создать переменную для использования в цикле
while k < 10: # условие для блока цикла
    print(k) # он будет печатать числа в столбце от 0 до 9
    k += 1 # увеличивает переменную 1. Когда k равно 10, цикл останавливается.
```  
            
**<span style="color:teal "> *While* - один из самых универсальных циклов в Python, поэтому он довольно медленный. Пришло время примера.</span>**

Посмотрите на пример в ячейке ниже.

**<span style="color:teal "> Учителю: пример простой. Мы продолжаем выводить число, пока оно меньше 100. Начальное число равно 10, и мы увеличиваем это число на 10 на каждой итерации. </span>**

````python
k = 10 # начальное значение для итераций
while k < 100: # условие для блока цикла
    print(k) # мы хотим получить столбец k-чисел 
    k += 10 # шаг k-чисел
# давайте проверим, что 40 должно быть последним числом из-за прерывания
````

**<span style = "color: Teal"> Одна из самых больших проблем молодых программистов - выучить правило «не повторяйся». Это означает, что вы не должны писать один и тот же код несколько раз. Когда вы это делаете, вы знаете, что часть кода должна перейти в функцию. </span>**
 
**<span style = "color: Teal"> Функция в Python - это объект, который принимает аргументы и возвращает данные. И он должен быть вызван для запуска. </span>**

*`функция`* - это блок кода, который запускается только при его вызове.

**<span style="color:teal "> Функция - это блок кода, который начинается с ключевого слова *def*, имени функции и двоеточия, например:</span>**


Как писать функции:
```python
def sum_of_2(x, y): # сделать функцию суммирования 2 переменных, они могут быть числами, строками или списками
    return x + y # чтобы сделать функцию, мы пишем: def название_функции(параметры): ... затем: return (результат)
``` 

**<span style="color:teal "> Каждая функция дает определенный результат. Если вы не укажете выдачу определенного результата в *return*, функция тем не менее выдаст результат *None* (ничего). В нашем примере мы указали, что нужно *вернуть* результат суммирования *x* и *y*. Как видите, мы можем вызвать функцию, передав два значения. Если вы передали недостаточное или слишком много аргументов для этой функции, вы получите сообщение об ошибке.</span>**
    

**<span style="color:teal "> Чтобы вызвать функцию, мы вводим имя функции и аргументы в скобках. </span>**

### <span style="color:Maroon"> <center> Критерии создания функций </span>

**<span style="color:teal "> Студенты, которые прошли курс основ Geeklama, должны быть с этим знакомы. Для остальных студентов это может быть новый материал. </span>**
    
1. __Функция должна быть многоразовой__ - используйте параметры для обобщения функции
2. __Функция выполняет только одно действие__ - одна функция должна выполнять одно действие, например, найти среднее значение. 
3. __Название функции состоит только из букв, цифр и знаков подчеркивания..__ Имя функции должно начинаться с буквы
4. __Функция должна включать комментарии__ которые объясняют цель функции и результат
    
**<span style="color:teal "> Сложнее всего понять, что функция должна делать только одно - это означает, что мы не должны помещать половину программы в одну функцию. Одна функция не должна вычислять среднее значение и находить среднее значение. Для этих двух действий нам нужны две отдельные функции. Это помогает с организацией кода. </span>**
    
В качестве примера рассмотрим функцию со списком в качестве аргумента, которая вернет новый список, состоящий из двойных элементов исходного списка.

**<span style="color:teal "> Обратите внимание на код и попросите учащихся ответить на вопросы для обсуждения. </span>**

````python
def Double(list1): 
    """Создает новый список, в котором дублируется каждый элемент"""
    list2 = [] # создать пустой список, который мы вернем с двойными элементами исходного списка
    for i in list1: # пройти через каждый элемент в начальном списке
        list2.append(i) # добавить элемент два раза в новый список
        list2.append(i)
    return list2 # в результате мы получим новый список с двойными элементами исходного списка

print(double([1,2,3,4,5,6])) # проверка с помощью print для вызова функции
````

````python
#Ваш код здесь

import arcade

def square(x,y,n): 
    """функция повторения квадратов"""
    for i in range(n):
        arcade.draw_rectangle_outline (x,y, i*10, i*10, arcade.csscolor.YELLOW)
    return print('квадраты готовы')

def circle(x,y,n): 
    """функция повторения кругов"""
    for i in range(n):
        arcade.draw_circle_outline (x, y, i*10, arcade.csscolor.RED)
    return print('круги готовы')
    
#открываем окно
arcade.open_window(700, 700, 'Pattern') # это стандартная система координат: (0;0) - левый нижний угол
# (700,700) - правый верхний угол

#задаем задний фон 
arcade.set_background_color(arcade.csscolor.SKY_BLUE)

arcade.start_render()

# отрисовываем шаблоны
square (300,300, 10)
circle(500,500,10)
square (100,100, 5)
circle(600,600,5)
square (400,400, 20)
circle(200,200,30)

arcade.finish_render()

#запустить окно, пока пользователь не закроет его
arcade.run()    

#
````

# **<span style="color:Maroon"> Дополнительная тема: операторы Break и continue**

**<span style="color:teal "> Вы можете использовать операторы *break* и *continue* внутри блока цикла. Давайте рассмотрим следующие объяснения их и того, как их использовать.</span>**

Внутри блока цикла довольно часто используются операторы *break* и *continue*.

**<span style="color:teal "> Учителю: прочтите следующие определения и объясните их двумя словами.</span>**

Это полезные выражение для использования внутри цикла:

- *`continue`* - оператор, который может остановить текущую итерацию и продолжить следующую.

  ```python
  for i in 'Остров':
        if i == 'о': # пройти через символы строки, проверяя, есть ли 'о'

            continue # он пропустит итерацию с символом 'a'
        print(i) # поскольку есть буква 'о', программа будет печатать 'стрв' каждым символом в своей строке
```

- *`break`* - оператор, который может остановить цикл до того, как он пройдёт через все элементы. 
  ```python
  for i in 'Остров':
        if i == 'a':  # пройти через символы строки, проверяя, есть ли 'о'
            break # он остановит цикл, если есть символ 'о'
        print(i) # так как есть буква 'о', поэтому программа будет печатать только 'стрв' каждым символом в своей строке
```  

Давайте рассмотрим небольшой пример списка.

**<span style = "color: teal"> Давайте попробуем применить оператор *continue* в *цикле* для списка. </span>**

**<span style = "color: teal"> Учителю: этот пример показывает, как *continue* работает в цикле. Вы должны показать, что *continue* пропускает итерацию, указанную в условии *if*. </span>**

````python
for i in ['Не','Солнечный','Остров']: # цикл for для списка
    if i == 'Не': # пройти через элементы, проверяя, есть ли «Нет»
        continue # он пропустит итерацию с element'Not '
    print(i) # так как есть элемент Not, поэтому давайте проверим, что программа напечатает
````
