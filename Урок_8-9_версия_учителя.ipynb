# __<span style = "color: Maroon"> Библиотека Arcade: ревизия </span>__

**<span style = "color: teal"> Ниже мы показываем код, с которого мы начнем. Чтобы сэкономить время, не будем начинать с нуля. Здесь нет ничего нового, кроме некоторых переменных, которые мы будем использовать при расширении нашего кода. Посмотрим, что запомнилось учащимся из этого материала. </span>**

Пришло время сделать игру графической. Мы начнем с кода, приведенного ниже. <br>
Посмотрите на приведенный ниже код, чтобы понять, что делает каждый метод.

````python
import arcade
import random

SPRITE_SCALING = 0.5
SCREEN_WIDTH = 450
SCREEN_HEIGHT = 500
SCREEN_TITLE = "Игра Приключения"
MOVEMENT_SPEED = 5


class MyGame(arcade.Window):
    """ Основной класс приложения """

    def __init__(self, width, height, title):
        """ Инициализатор """
        super().__init__(width, height, title)

        # Переменные для комнат
        self.current_room = 0
        self.rooms = None

        # Переменные для игрока
        self.player_sprite = None
        self.player_list = None

        # Другие переменные
        self.win = False
        self.counter = 0

    def setup(self):
        """ Настраиваем игру и инициализируем переменные """
        # Настроиваем игрока
        self.player_sprite = arcade.Sprite(":resources:images/animated_characters/female_person/femalePerson_idle.png",
                                           SPRITE_SCALING)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list = arcade.SpriteList()
        self.player_list.append(self.player_sprite)

        # Наш список комнат
        self.rooms = []

    def on_draw(self):
        """Метод отрисовки"""
        arcade.start_render()
        self.player_list.draw()

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажата клавиша """
        if key == arcade.key.UP:
            self.player_sprite.change_y = MOVEMENT_SPEED
        elif key == arcade.key.DOWN:
            self.player_sprite.change_y = -MOVEMENT_SPEED
        elif key == arcade.key.LEFT:
            self.player_sprite.change_x = -MOVEMENT_SPEED
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = MOVEMENT_SPEED

    def on_key_release(self, key, modifiers):
        """Вызывается, когда пользователь отпускает клавишу """
        if key == arcade.key.UP or key == arcade.key.DOWN:
            self.player_sprite.change_y = 0
        elif key == arcade.key.LEFT or key == arcade.key.RIGHT:
            self.player_sprite.change_x = 0

    def keep_walls(self):
        """Этот метод не дает игроку выйти за пределы экрана"""
        if self.player_sprite.center_x < 0 or self.player_sprite.center_x > SCREEN_WIDTH:
            self.player_sprite.change_x = 0
        elif self.player_sprite.center_y < 0 or self.player_sprite.center_y > SCREEN_HEIGHT:
            self.player_sprite.change_y = 0

    def update(self, delta_time):
        """ Движения и игровая логика """
        self.player_list.update()
        self.keep_walls()


def main():
    """ Метод main """
    window = MyGame(SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_TITLE)
    window.setup()
    arcade.run()


main()
````

````python
import arcade
import random

SPRITE_SCALING = 0.7
SCREEN_WIDTH = 500
SCREEN_HEIGHT = 500
SCREEN_TITLE = "Игра Приключения"
MOVEMENT_SPEED = 5


class Room:
    """Этот класс содержит всю информацию о разных комнатах."""

    def __init__(self):
        # Вам может понадобиться много списков. Списки монет, монстров и т.д.
        self.coins_list = None
        # Фоновые изображения.
        self.background = None


def setup_room_0():
    """Создать и вернуть комнату 0."""
    room = Room()
    room.background = arcade.load_texture("./Images/hallway.jpg")
    return room


def setup_room_1():
    """Создать и вернуть комнату 1."""
    room = Room()
    room.background = arcade.load_texture("./Images/room1.jpg")
    return room


def setup_room_2():
    """Создать и вернуть комнату 2."""
    room = Room()
    room.background = arcade.load_texture("./Images/hallway.jpg")
    return room


def setup_room_3():
    """Создать и вернуть комнату 3. """
    room = Room()
    room.background = arcade.load_texture("./Images/room3.jpg")
    return room


def setup_room_4():
    """Создать и вернуть комнату 4."""
    room = Room()
    room.background = arcade.load_texture("./Images/room4.jpg")
    return room


class MyGame(arcade.Window):
    """ Основной класс приложения """

    def __init__(self, width, height, title):
        """ Инициализатор """
        super().__init__(width, height, title)

        # Переменные для комнат
        self.current_room = 0
        self.rooms = None

        # Переменные для игрока
        self.player_sprite = None
        self.player_list = None

        # Другие переменные
        self.win = False
        self.counter = 0

    def setup(self):
        """ Настраиваем игру и инициализируем переменные """
        # Настроиваем игрока
        self.player_sprite = arcade.Sprite(":resources:images/animated_characters/female_person/femalePerson_idle.png",
                                           SPRITE_SCALING)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list = arcade.SpriteList()
        self.player_list.append(self.player_sprite)

        # Наш список комнат
        self.rooms = []
        self.rooms.append(setup_room_0())
        self.rooms.append(setup_room_1())
        self.rooms.append(setup_room_2())
        self.rooms.append(setup_room_3())
        self.rooms.append(setup_room_4())

    def on_draw(self):
        """Метод отрисовки"""
        arcade.start_render()

        # Отрисовать задний фон
        arcade.draw_lrwh_rectangle_textured(0, 0, SCREEN_WIDTH, SCREEN_HEIGHT, self.rooms[self.current_room].background)

        self.player_list.draw()

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажата клавиша """
        if key == arcade.key.UP:
            self.player_sprite.change_y = MOVEMENT_SPEED
        elif key == arcade.key.DOWN:
            self.player_sprite.change_y = -MOVEMENT_SPEED
        elif key == arcade.key.LEFT:
            self.player_sprite.change_x = -MOVEMENT_SPEED
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = MOVEMENT_SPEED

    def on_key_release(self, key, modifiers):
        """Вызывается, когда пользователь отпускает клавишу """
        if key == arcade.key.UP or key == arcade.key.DOWN:
            self.player_sprite.change_y = 0
        elif key == arcade.key.LEFT or key == arcade.key.RIGHT:
            self.player_sprite.change_x = 0

    def keep_walls(self):
        """Этот метод не дает игроку выйти за пределы экрана"""
        if self.player_sprite.center_x < 0 or self.player_sprite.center_x > SCREEN_WIDTH:
            self.player_sprite.change_x = 0
        elif self.player_sprite.center_y < 0 or self.player_sprite.center_y > SCREEN_HEIGHT:
            self.player_sprite.change_y = 0

    def update(self, delta_time):
        """ Движения и игровая логика """
        self.player_list.update()
        self.keep_walls()


def main():
    """ Метод main """
    window = MyGame(SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_TITLE)
    window.setup()
    arcade.run()


if __name__ == "__main__":
    main()
````
