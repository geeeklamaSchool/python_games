````python
#Ширина и высота устанавливаются в соответствии с размером изображения
import arcade
SCREEN_WIDTH = 750 
SCREEN_HEIGHT = 600

class Game(arcade.Window):

    def __init__(self, width, height, title):

        # Вызвать функцию инициализации родительского класса
        super().__init__(width, height, title)
        
        # Загрузить текстуры
        self.background = arcade.load_texture("./Images/water.jpg") #загрузить изображение заднего фона

    def on_draw(self):
        """ Вызывается всякий раз, когда нам нужно нарисовать окно """
        arcade.start_render()
        #задаем изображение для заднего фона
        arcade.draw_lrwh_rectangle_textured(0, 0,SCREEN_WIDTH, SCREEN_HEIGHT,self.background) 
        

def main():
    window = Game(SCREEN_WIDTH, SCREEN_HEIGHT, "Fish game")

    arcade.run()

main()
````

## <center> <span style = "color: goldenrod"> Что вы думаете? </span>
__<center> Как мы можем решить данную проблему выхода рыбы за пределы экрана (какая логика нам здесь нужна)? <br>__

Посмотрим, насколько вы были близки к правде. Обратите внимание на код ниже.
    
__<span style = "color: midnightblue"> P.S. </span>__ Обратите внимание на комментарии, они помогают понять логику кода.

````python
import arcade
#Это глобальные переменные 
SCREEN_WIDTH = 750
SCREEN_HEIGHT = 600


class Game(arcade.Window):

    def __init__(self, width, height, title):

        # Вызываем функцию инициализации родительского класса
        super().__init__(width, height, title)

        # Загружаем текстуры
        self.background = arcade.load_texture("./Images/water.jpg") #загружаем изображение заднего фона
        self.fish1 = arcade.load_texture("./Images/fish1.png")
        
        self.x_pos = 400 # позиция объекта по оси x
        self.y_pos = 200 # # позиция объекта по оси x
        self.size = 70 # размер объекта
        self.move_x = 1  # Скорость движения по оси x (с которой мы обновляем нашу позицию)
        self.move_y = 1  # Скорость движения по оси y (с которой мы обновляем нашу позицию)

    def on_draw(self):
        """ Вызывается всякий раз, когда нам нужно нарисовать окно. """
        arcade.start_render()

        arcade.draw_lrwh_rectangle_textured(0, 0,SCREEN_WIDTH, SCREEN_HEIGHT,self.background)
        arcade.draw_lrwh_rectangle_textured(self.x_pos, self.y_pos, self.size, self.size, self.fish1)
        
    def update(self, delta_time):
        """ Вызывается, чтобы обновить наши объекты (примерно 60 раз в секунду)"""
        self.x_pos += self.move_x 
        self.y_pos += self.move_y
        
        # Когда квадрат достигает края экрана, он меняет направление.
        if self.x_pos < 0: #развород при достижении левой стороны (координата x здесь равна 0)
            self.move_x *= -1

        if self.x_pos > SCREEN_WIDTH - self.size: #разворот при достижении правой стороны
            self.move_x *= -1
        
        if self.y_pos < 0: #разворот при достижении нижней стороны (координата y здесь равна 0)
            self.move_y *= -1

        if self.y_pos > SCREEN_HEIGHT - self.size: #разворот при достижении верхней стороны
            self.move_y *= -1
            
    
def main():
    window = Game(SCREEN_WIDTH, SCREEN_HEIGHT, "Fish game")
    arcade.run()

main()

````

Важно, чтобы учащиеся понимали логику кода. <br> Обратите внимание, что `*=` имеет аналогичную логику, что и `+=`, только переменная умножается на заданное число и обновляется (вместо добавления к числу). <br> Отметьте важность параметризации и комментариев. **<span style = "color: teal"> Теперь наша рыба движется сама по себе. Было бы интереснее позволить пользователю управлять рыбой. Мы уже знаем два метода в _arcade.Window_, которые мы можем использовать для взаимодействия с пользователем: _on_key_press_ и _on_key_release_. Если мы хотим заставить рыбу двигаться на основе управления клавиатурой, нам нужно обновлять положение только при нажатии клавиши. Когда клавиша отпускается, позиция не изменяется (изменяется на 0). Метод _update_ вызывается 60 раз в секунду - мы не можем его контролировать. Но мы можем управлять переменной, которая сообщает, насколько изменилась позиция. Есть еще два метода, которые мы можем использовать для получения пользовательского ввода с помощью мыши: _on_mouse_motion_ и _on_mouse_press_ <br> </span>**

Способы пользовательского контроля:
1. `on_key_press` - вызывается при нажатии клавиши
2. `on_key_release` - вызывается при отпускании клавиши
3. `on_mouse_motion` - вызывается при перемещении мыши
4. `on_mouse_press` - вызывается при нажатии кнопки мыши

Мы знакомы с методами взаимодействия пользователя с клавиатурой. Обратите внимание на приведенный ниже код, чтобы увидеть, как они реализованы вместе с методом _update_.

**<span style = "color: teal"> Напомните, что у нас есть глобальная переменная SPEED - обозначает скорость, с которой движется наша рыба. Когда конкретная клавиша нажата, move_x или move_y устанавливаются с положительной или отрицательной скоростью. Знак обозначает направление движения. Когда клавиша отпускается, move_x / move_y устанавливаются в 0. <br> Также обратите внимание, что теперь мы создаем объект рыбы с move_x и move_y равными нулю, поэтому она не перемещается, пока пользователь не нажмет клавишу. </span>**

````python
import arcade
#Это глобальные переменные 
SCREEN_WIDTH = 750
SCREEN_HEIGHT = 600
SPEED = 3

class Fish:

    def __init__(self, x_pos, y_pos, size, move_x, move_y, texture):
'''Этот метод инициализирует информацию о квадрате'''        
        
        self.fish1 = arcade.load_texture(texture)
        
        self.x_pos = x_pos
        self.y_pos = y_pos
        self.size = size
        self.move_x = move_x 
        self.move_y = move_y

    def draw_fish(self):
        """ отрисовывает квадрат с данными переменными """
        arcade.draw_lrwh_rectangle_textured(self.x_pos, self.y_pos, self.size, self.size, self.fish1)
        
    def update(self):
        """ Вызывается, чтобы обновить наши объекты (примерно 60 раз в секунду)"""
        self.x_pos += self.move_x 
        self.y_pos += self.move_y
        
        # Когда квадрат достигает края экрана, он меняет направление.
        if self.x_pos < 0: #развород при достижении левой стороны (координата x здесь равна 0)
            self.move_x *= -1

        if self.x_pos > SCREEN_WIDTH - self.size: #разворот при достижении правой стороны
            self.move_x *= -1
        
        if self.y_pos < 0: #разворот при достижении нижней стороны (координата y здесь равна 0)
            self.move_y *= -1

        if self.y_pos > SCREEN_HEIGHT - self.size: #разворот при достижении верхней стороны
            self.move_y *= -1

class Game(arcade.Window):

    def __init__(self, width, height, title):

        # Вызываем функцию инициализации родительского класса
        super().__init__(width, height, title)

        # Загружаем текстуры
        self.background = arcade.load_texture("./Images/water.jpg") #load the backgroung image
        
        self.fish1 = Fish(400,200,70,0,0,"./Images/fish1.png")

    def on_draw(self):
        """ Вызывается всякий раз, когда нам нужно нарисовать окно. """
        arcade.start_render()

        arcade.draw_lrwh_rectangle_textured(0, 0,SCREEN_WIDTH, SCREEN_HEIGHT,self.background)
        
        self.fish1.draw_fish()
        
    def update(self, delta_time):
        """ Вызывается, чтобы обновить наши объекты (примерно 60 раз в секунду)"""
        self.fish1.update()

        
    def on_key_press(self, key, modifiers):
        """ Вызывается всякий раз, когда пользователь нажимает кнопку """
        if key == arcade.key.LEFT:
            self.fish1.move_x = -SPEED
        elif key == arcade.key.RIGHT:
            self.fish1.move_x = SPEED
        elif key == arcade.key.UP:
            self.fish1.move_y = SPEED
        elif key == arcade.key.DOWN:
            self.fish1.move_y = -SPEED

    def on_key_release(self, key, modifiers):
        """ Вызывается всякий раз, когда пользователь отпускает кнопку """
        if key == arcade.key.LEFT or key == arcade.key.RIGHT:
            self.fish1.move_x = 0
        elif key == arcade.key.UP or key == arcade.key.DOWN:
            self.fish1.move_y = 0        
    
def main():
    window = Game(SCREEN_WIDTH, SCREEN_HEIGHT, "Fish game")
    arcade.run()

main()
````

````python
import arcade

SCREEN_WIDTH = 750
SCREEN_HEIGHT = 600
MOVEMENT_SPEED = 3


class Cursor:
    def __init__(self, position_x, position_y, change_x, change_y, size):

        # Берем параметры функции инициализации выше 
        # и создаем из них переменные экземпляра.        
        self.position_x = position_x
        self.position_y = position_y
        self.change_x = change_x
        self.change_y = change_y
        self.size = size
        
        self.cursor = arcade.load_texture("./Images/cursor.png")
    

    def draw(self):
        """ Отрисовываем курсор с имеющимися у нас переменными экземпляра. """
        arcade.draw_lrwh_rectangle_textured(self.position_x, self.position_y, self.size, self.size, self.cursor)

    def update(self):
        # Двигаем курсор
        self.position_y += self.change_y
        self.position_x += self.change_x

        # Посмотрите, попал ли курсор в край экрана. Если да, измените направление        
        if self.position_x < 0:
            self.position_x = 0

        if self.position_x > SCREEN_WIDTH:
            self.position_x = SCREEN_WIDTH

        if self.position_y < 0:
            self.position_y = 0

        if self.position_y > SCREEN_HEIGHT:
            self.position_y = SCREEN_HEIGHT

````

````python

class MyGame(arcade.Window):

    def __init__(self, width, height, title):

        # Вызываем функцию инициализации родительского класса
        super().__init__(width, height, title)

        # Курсор мыши не должен быть виден, когда будет над окном
        # Чтобы мы видели только объект, без курсора
        self.set_mouse_visible(False)

        #задаем текстуру заднего фона
        self.background = arcade.load_texture("./Images/island_map.jpg")
        #создаем текступу сокровища
        self.treasure = arcade.load_texture("./Images/treasure.png")
        
        # создаем наш курсор
        self.cursor = Cursor(50, 50, 0, 0, 70)
        self.counter = 0
        self.var = 0
          

    def on_draw(self):
        """ Вызывается всякий раз, когда нам нужно отрисовать окно """
        arcade.start_render()
        arcade.draw_lrwh_rectangle_textured(0, 0,SCREEN_WIDTH, SCREEN_HEIGHT,self.background)
        self.cursor.draw()
        
        if 50 < self.cursor.position_x < 80 and 100 < self.cursor.position_y < 200:
            arcade.draw_text ('Найдено', 320,340, arcade.csscolor.YELLOW, font_name = 'calibri', bold = True, font_size = 30)
        if (self.counter == 1 and 50 < self.cursor.position_x < 80 and 100 < self.cursor.position_y < 200) or self.var>=1:
            arcade.draw_lrwh_rectangle_textured(320, 340, 150, 75,self.treasure)
            self.var+=1
                         

    def update(self, delta_time):
        self.cursor.update()
        
    def on_key_press(self, key, modifiers):
        """ Вызывается всякий раз, когда пользователь нажимает кнопку """
        if key == arcade.key.ENTER:
            self.counter = 1

    def on_key_release(self, key, modifiers):
        """ Вызывается всякий раз, когда пользователь отпускает кнопку """
        if key == arcade.key.ENTER:
            self.counter = 0

            
    def on_mouse_motion(self, x, y, dx, dy):
        """ Вызывается, чтобы обновить наши объекты (примерно 60 раз в секунду)"""
        self.cursor.position_x = x
        self.cursor.position_y = y
````

````python
def main():
    window = MyGame(SCREEN_WIDTH, SCREEN_HEIGHT, "Пример отрисовки")
    arcade.run()

main()
````
