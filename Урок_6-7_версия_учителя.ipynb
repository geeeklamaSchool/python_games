````python
import arcade
import random

SCREEN_WIDTH = 725
SCREEN_HEIGHT = 607
number_coins = 50
number_enemies = 30


class MyGame(arcade.Window):
    """ Наш специальный класс Window """

    def __init__(self):
        """ Инициализатор """
        # Вызываем инициализатор родительского класса
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, "Битва на острове")

        # Отключаем видимость курсора
        self.set_mouse_visible(False)

        # Создаем текстуру заднего фона
        self.background = arcade.load_texture("./Images/water.jpg")

        # Переменные для списков спрайтов
        self.player_list = arcade.SpriteList()
        self.coin_list = arcade.SpriteList()
        self.enemies_list = arcade.SpriteList()

        # Счетчик монет
        self.counter = 0

    def gen_random(self, number, scale, image_path, alist, direction):
        """ Метод, который создает спрайты в случайных местах и добавляет их в список """
        for i in range(number):
            sprite_name = arcade.Sprite(image_path, scale)
            sprite_name.center_x = random.randrange(SCREEN_WIDTH)
            sprite_name.center_y = random.randrange(SCREEN_HEIGHT)
            sprite_name.velocity = (0, direction * random.randrange(1, 5))
            alist.append(sprite_name)

    def setup(self):
        """ Настраиваем игру и инициализируем переменные. """

        # Задаем спрайт игрока
        self.player_sprite = arcade.Sprite(":resources:images/enemies/fishGreen.png", 0.5)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list.append(self.player_sprite)

        # Задаем спрайты монет
        self.gen_random(number_coins, 0.3, ":resources:images/items/coinGold.png", self.coin_list, -1)

        # Задаем спрайты врагов
        self.gen_random(number_enemies, 0.4, ":resources:images/enemies/slimeBlue.png", self.enemies_list, 1)

    def on_draw(self):
        """ Этот метод используется для отрисовки """
        arcade.start_render()
        arcade.draw_lrwh_rectangle_textured(0, 0, SCREEN_WIDTH, SCREEN_HEIGHT, self.background)

        # Отрисовываем спрайты
        self.player_list.draw()
        self.coin_list.draw()
        self.enemies_list.draw()

        # Отображаем счетчик очков на экране
        output = f"Счет: {self.counter}"
        arcade.draw_text(output, 10, 20, arcade.color.WHITE, 14)

    def update(self, delta_time):
        """ Анимация и логика """

        # Вызов обновления всех спрайтов
        self.coin_list.update()
        self.enemies_list.update()

        for i in self.coin_list:
            if i.center_y < 0:
                i.center_y = SCREEN_HEIGHT

        for i in self.enemies_list:
            if i.center_y > SCREEN_HEIGHT:
                i.center_y = 0

        # Взаимодействия игрока и монет
        coins_hit_list = self.player_sprite.collides_with_list(self.coin_list)

        for i in coins_hit_list:
            i.remove_from_sprite_lists()  # метод, который удаляет прайт из списка
            self.counter += 1

        # Взаимодействия игрока и врагов
        enemies_hit_list = self.player_sprite.collides_with_list(self.enemies_list)

        if len(enemies_hit_list) > 0:
            self.counter = 0

        # Как только все монеты будут собраны, все враги исчезают
        if len(self.coin_list) == 0:
            for i in self.enemies_list:
                i.remove_from_sprite_lists()

    def on_mouse_motion(self, x, y, dx, dy):
        """ Обработка движения мыши """
        # Перемещаем центр спрайта игрока в соответствии с положением мыши по осям x и y.
        self.player_sprite.center_x = x
        self.player_sprite.center_y = y


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()


main()
````

````python
import arcade
import random

SCREEN_WIDTH = 725
SCREEN_HEIGHT = 607
number_coins = 50
number_enemies = 30


class MyGame(arcade.Window):
    """ Наш специальный класс Window """

    def __init__(self):
        """ Инициализатор """
        # Вызываем инициализатор родительского класса
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, "Битва на острове")

        # Отключаем видимость курсора
        self.set_mouse_visible(False)

        # Создаем текстуру заднего фона
        self.background = arcade.load_texture("./Images/water.jpg")

        # Переменные для списков спрайтов
        self.player_list = arcade.SpriteList()
        self.coin_list = arcade.SpriteList()
        self.enemies_list = arcade.SpriteList()
        self.bullet_list = arcade.SpriteList()

        # Счетчик монет
        self.counter = 0

        # Загрузка звуков
        self.coin_sound = arcade.load_sound(":resources:sounds/coin3.wav")
        self.enemy_sound = arcade.load_sound(":resources:sounds/hurt4.wav")
        self.bullet_sound = arcade.load_sound(":resources:sounds/hit5.wav")
        self.hit_sound = arcade.load_sound(":resources:sounds/error4.wav")

    def gen_random(self, number, scale, image_path, alist, direction):
        """ Метод, который создает спрайты в случайных местах и добавляет их в список """
        for i in range(number):
            sprite_name = arcade.Sprite(image_path, scale)
            sprite_name.center_x = random.randrange(SCREEN_WIDTH)
            sprite_name.center_y = random.randrange(SCREEN_HEIGHT)
            sprite_name.velocity = (0, direction * random.randrange(1, 5))
            alist.append(sprite_name)

    def setup(self):
        """ Настраиваем игру и инициализируем переменные. """

        # Задаем спрайт игрока
        self.player_sprite = arcade.Sprite(":resources:images/enemies/fishGreen.png", 0.5)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list.append(self.player_sprite)

        # Задаем спрайты монет
        self.gen_random(number_coins, 0.3, ":resources:images/items/coinGold.png", self.coin_list, -1)

        # Задаем спрайты врагов
        self.gen_random(number_enemies, 0.4, ":resources:images/enemies/slimeBlue.png", self.enemies_list, 1)

    def on_draw(self):
        """ Этот метод используется для отрисовки """
        arcade.start_render()
        arcade.draw_lrwh_rectangle_textured(0, 0, SCREEN_WIDTH, SCREEN_HEIGHT, self.background)

        # Отрисовываем спрайты
        self.player_list.draw()
        self.coin_list.draw()
        self.enemies_list.draw()
        self.bullet_list.draw()

        # Отображаем счетчик очков на экране
        output = f"Счет: {self.counter}"
        arcade.draw_text(output, 10, 20, arcade.color.WHITE, 14)

    def sprite_recycle(self):
        """Этот метод заставляет спрайты появляться на противоположной стороне экрана,
         когда они достигают одной стороны."""
        for i in self.coin_list:
            if i.center_y < 0:
                i.center_y = SCREEN_HEIGHT

        for i in self.enemies_list:
            if i.center_y > SCREEN_HEIGHT:
                i.center_y = 0

    def coin_collect(self):
        """Взаимодействие игрока и монет"""
        coins_hit_list = self.player_sprite.collides_with_list(self.coin_list)

        for i in coins_hit_list:
            i.remove_from_sprite_lists()  # метод, который удаляет спрайт из списка
            self.counter += 1
            arcade.play_sound(self.coin_sound)  # звук сбора монеты

    def enemy_interactions(self):
        """Взаимодействие игрока и врагов"""
        enemies_hit_list = self.player_sprite.collides_with_list(self.enemies_list)

        if len(enemies_hit_list) > 0:
            self.counter = 0
            arcade.play_sound(self.enemy_sound)

        # Как только все монеты будут собраны, все враги исчезают
        if len(self.coin_list) == 0:
            for i in self.enemies_list:
                i.remove_from_sprite_lists()

    def bullet_interactions(self):
        """Взаимодействие пуль с врагами"""
        for i in self.bullet_list:
            enemies_hit_list = i.collides_with_list(self.enemies_list)
            # удалить врагов, в которых попала пуля
            for j in enemies_hit_list:
                j.remove_from_sprite_lists()
            # удалить пулю
            if len(enemies_hit_list) > 0:
                i.remove_from_sprite_lists()
                arcade.play_sound(self.hit_sound)  # звук столкновения пули и врага

    def update(self, delta_time):
        """ Анимация и логика """
        # Вызвать методы обновления всех спрайтов
        self.coin_list.update()
        self.enemies_list.update()
        self.bullet_list.update()
        self.sprite_recycle()
        self.coin_collect()
        self.enemy_interactions()
        self.bullet_interactions()

    def on_mouse_motion(self, x, y, dx, dy):
        """ Обработка движения мыши """
        # Перемещаем центр спрайта игрока в соответствии с положением мыши по осям x и y.
        self.player_sprite.center_x = x
        self.player_sprite.center_y = y

    def on_mouse_press(self, x, y, button, modifiers):
        """ Обработка нажатий мыши """
        # если левая кнопка нажата, создать пулю
        if button == arcade.MOUSE_BUTTON_LEFT:
            bullet = arcade.Sprite(":resources:images/space_shooter/meteorGrey_tiny1.png", 0.5)
            bullet.center_x = x
            bullet.center_y = y
            bullet.velocity = (0, 10)
            self.bullet_list.append(bullet)
            arcade.play_sound(self.bullet_sound)  # sound for bullet shooting


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()


main()
````

````python
import arcade
import random

SCREEN_WIDTH = 725
SCREEN_HEIGHT = 607
number_coins = 50
number_enemies = 30


class MyGame(arcade.Window):
    """ Наш специальный класс Window """

    def __init__(self):
        """ Инициализатор """
        # Вызываем инициализатор родительского класса
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, "Битва на острове")

        # Отключаем видимость курсора
        self.set_mouse_visible(False)

        # Создаем текстуру заднего фона
        self.background = arcade.load_texture("./Images/water.jpg")

        # Переменные для списков спрайтов
        self.player_list = arcade.SpriteList()
        self.coin_list = arcade.SpriteList()
        self.enemies_list = arcade.SpriteList()
        self.bullet_list = arcade.SpriteList()

        # Счетчик монет
        self.counter = 0

        # Загрузка звуков
        self.coin_sound = arcade.load_sound(":resources:sounds/coin3.wav")
        self.enemy_sound = arcade.load_sound(":resources:sounds/hurt4.wav")
        self.bullet_sound = arcade.load_sound(":resources:sounds/hit5.wav")
        self.hit_sound = arcade.load_sound(":resources:sounds/error4.wav")

        # изначально игра не на паузе
        self.paused = False
        # флаг выхода из игры
        self.quit = False
        self.resume = False
        self.enter_state = False

    def gen_random(self, number, scale, image_path, alist, direction):
        """ Метод, который создает спрайты в случайных местах и добавляет их в список """
        for i in range(number):
            sprite_name = arcade.Sprite(image_path, scale)
            sprite_name.center_x = random.randrange(SCREEN_WIDTH)
            sprite_name.center_y = random.randrange(SCREEN_HEIGHT)
            sprite_name.velocity = (0, direction * random.randrange(1, 5))
            alist.append(sprite_name)

    def setup(self):
        """ Настраиваем игру и инициализируем переменные. """

        # Задаем спрайт игрока
        self.player_sprite = arcade.Sprite(":resources:images/enemies/fishGreen.png", 0.5)
        self.player_sprite.center_x = 100
        self.player_sprite.center_y = 100
        self.player_list.append(self.player_sprite)

        # Задаем спрайты монет
        self.gen_random(number_coins, 0.3, ":resources:images/items/coinGold.png", self.coin_list, -1)

        # Задаем спрайты врагов
        self.gen_random(number_enemies, 0.4, ":resources:images/enemies/slimeBlue.png", self.enemies_list, 1)

    def on_draw(self):
        """ Этот метод используется для отрисовки """
        arcade.start_render()
        arcade.draw_lrwh_rectangle_textured(0, 0, SCREEN_WIDTH, SCREEN_HEIGHT, self.background)

        # Отрисовываем спрайты
        self.player_list.draw()
        self.coin_list.draw()
        self.enemies_list.draw()
        self.bullet_list.draw()

        # Отображаем счетчик очков на экране
        output = f"Счет: {self.counter}"
        arcade.draw_text(output, 10, 20, arcade.color.WHITE, 14)

        # отобразить сообщения
        if self.paused:
            arcade.draw_text("ПАУЗА", 30, SCREEN_HEIGHT / 2, arcade.color.WHITE, 120)
        if self.quit and self.resume:
            arcade.draw_text("Вы действительно хотите выйти из игры?", 20, 200, arcade.color.BLACK, 35)
            arcade.draw_text("Нажмите Enter, чтобы выйти. Нажмите R, чтобы возобновить игру.", 20, 150, arcade.color.BLACK, 25)

    def sprite_recycle(self):
        """Этот метод заставляет спрайты появляться на противоположной стороне экрана,
         когда они достигают одной стороны."""
        for i in self.coin_list:
            if i.center_y < 0:
                i.center_y = SCREEN_HEIGHT

        for i in self.enemies_list:
            if i.center_y > SCREEN_HEIGHT:
                i.center_y = 0

    def coin_collect(self):
        """Взаимодействие игрока и монет"""
        coins_hit_list = self.player_sprite.collides_with_list(self.coin_list)

        for i in coins_hit_list:
            i.remove_from_sprite_lists()  # метод, который удаляет спрайт из списка
            self.counter += 1
            arcade.play_sound(self.coin_sound)  # звук сбора монеты

    def enemy_interactions(self):
        """Взаимодействие игрока и врагов"""
        enemies_hit_list = self.player_sprite.collides_with_list(self.enemies_list)

        if len(enemies_hit_list) > 0:
            self.counter = 0
            arcade.play_sound(self.enemy_sound)

        # Как только все монеты будут собраны, все враги исчезают
        if len(self.coin_list) == 0:
            for i in self.enemies_list:
                i.remove_from_sprite_lists()

    def bullet_interactions(self):
        """Взаимодействие пуль с врагами"""
        for i in self.bullet_list:
            enemies_hit_list = i.collides_with_list(self.enemies_list)
            # удалить врагов, в которых попала пуля
            for j in enemies_hit_list:
                j.remove_from_sprite_lists()
            # удалить пулю
            if len(enemies_hit_list) > 0:
                i.remove_from_sprite_lists()
                arcade.play_sound(self.hit_sound)  # звук столкновения пули и врага

    def update(self, delta_time):
        """ Анимация и логика """
        
        if self.enter_state and self.quit:
            arcade.close_window()
        # Никаких обновлений на паузе
        if self.paused:
            return

        # Вызвать методы обновления всех спрайтов
        self.coin_list.update()
        self.enemies_list.update()
        self.bullet_list.update()
        self.sprite_recycle()
        self.coin_collect()
        self.enemy_interactions()
        self.bullet_interactions()

    def on_mouse_motion(self, x, y, dx, dy):
        """ Обработка движения мыши """
        # Перемещаем центр спрайта игрока в соответствии с положением мыши по осям x и y.
        self.player_sprite.center_x = x
        self.player_sprite.center_y = y

    def on_mouse_press(self, x, y, button, modifiers):
        """ Обработка нажатий мыши """
        # если левая кнопка нажата, создать пулю
        if button == arcade.MOUSE_BUTTON_LEFT and self.paused == False:
            bullet = arcade.Sprite(":resources:images/space_shooter/meteorGrey_tiny1.png", 0.5)
            bullet.center_x = x
            bullet.center_y = y
            bullet.velocity = (0, 10)
            self.bullet_list.append(bullet)
            arcade.play_sound(self.bullet_sound)  # sound for bullet shooting

    def on_key_press(self, key, modifiers):
        """ Обработка нажатий клавиш клавиатуры """
        # Поставить игру на паузу
        if key == arcade.key.P:
            self.paused = not self.paused
        # Выйти из игры
        if key == arcade.key.Q:
            self.paused = True
            self.quit = True
            self.resume = True
        if key == arcade.key.R:
            self.paused = False
            self.quit = False
            self.resume = False
        if key == arcade.key.ENTER:
            self.enter_state = not self.enter_state

    def on_key_release(self, key, modifiers):
        """ Обработка нажатий клавиатуры """
        # Установить игру на паузу
        if key == arcade.key.ENTER:
            self.enter_state = not self.enter_state


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()


main()
````
