## __<span style = "color: Maroon"> Рабочий процесс уроков на протяжении хакатона </span>__
    
Во время урока каждый ученик показывает экран и рассказывает, что сделано сегодня. Время, которое получает каждый ученик, зависит от размера группы. Время делится между всеми поровну. Учитель заранее сообщит, сколько времени есть у каждого ученика. Каждая презентация состоит из трех частей:
- **Презентация проекта** - объясните общую идею игры вместе с кодом. Отметьте интересные функции, которые вы добавили, и покажите, как вы реализовали их в коде.
- **Вопросы аудитории** - Учитель и другие ученики могут задавать вопросы и давать идеи/советы относительно кода и игры в целом.
- **Вопросы докладчика** - это ваша возможность попросить учителя (или одноклассников) помочь с кодом (например, как реализовать некоторую логику или как исправить некоторую ошибку).
