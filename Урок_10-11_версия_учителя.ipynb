````python
import arcade

# Константы
SCREEN_WIDTH = 1100
SCREEN_HEIGHT = 700
SCREEN_TITLE = "Wall arcade"


class MyGame(arcade.Window):
    """Игровой класс"""

    def __init__(self):
        # Вызвать родительский класс и настроить окно
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_TITLE)

        # Списки спрайтов
        self.player_list = None
        self.coin_list = None
        self.wall_list = None
        self.enemy_list = None

        # Отдельная переменная для спрайта игрока
        self.player_sprite = None

        # Загрузить изображение заднего фона
        self.background = arcade.load_texture("./Images/air.jpg")

    def setup(self):
        """ Настраиваем игру здесь. Вызываем эту функцию, чтобы перезапустить игру. """
        # Создать списки спрайтов
        self.player_list = arcade.SpriteList()

        # Настройте игрока, специально поместив его в эти координаты.
        self.player_sprite = arcade.Sprite("./Images/lama_left.png", 0.4)
        self.player_sprite.center_x = 64
        self.player_sprite.center_y = 115
        self.player_list.append(self.player_sprite)

    def on_draw(self):
        """ Метод отрисовки """
        arcade.start_render()

        # Отрисовать задний фон
        arcade.draw_lrwh_rectangle_textured(0, 0, SCREEN_WIDTH * 3, SCREEN_HEIGHT, self.background)

        # Отрисовать спрайты
        self.player_list.draw()


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()


main()
````

# __<span style = "color: Maroon"> Стены </span>__

**<span style = "color: teal"> В игре-платформере стеной считается все, через что игрок не может пройти. Стена также является спрайтом, как и монеты, игроки и враги. Вы можете выбрать любую картинку для стены. Мы будем использовать изображения из ресурсов библиотеки аркад. </span>**

**<span style = "color: teal"> Давайте добавим несколько стен в нашу игру. Вначале игроку нужен этаж. Мы также создадим стены, по которым игрок может прыгать, чтобы двигаться дальше. В размещении стены очень много кода </span>**

````python
import arcade

# Константы
SCREEN_WIDTH = 1100
SCREEN_HEIGHT = 700
SCREEN_TITLE = "Wall arcade"

SPEED = 5


class MyGame(arcade.Window):
    """Игровой класс"""

    def __init__(self):

        # Вызвать родительский класс и настроить окно
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_TITLE)

        # Списки спрайтов
        self.player_list = None
        self.coin_list = None
        self.wall_list = None
        self.enemy_list = None

        # Отдельная переменная для спрайта игрока
        self.player_sprite = None

        # Загрузить изображение заднего фона
        self.background = arcade.load_texture("./Images/air.jpg")

    def ground_setup(self, y, x_start, x_end, pic):
        """Создать пол (много стен в одну линию)"""
        for i in range(x_start, x_end, 64):
            wall = arcade.Sprite(pic, 0.5)
            wall.center_x = i
            wall.center_y = y
            self.wall_list.append(wall)

    def boxes_on_coords(self, alist, pic):
        """Создать стены по координатам, указанным в переданном списке"""
        for i in alist:
            # Добавить квадрат на пол
            wall = arcade.Sprite(pic, 0.5)
            wall.position = i
            self.wall_list.append(wall)

    def setup(self):
        """ Настраиваем игру здесь. Вызываем эту функцию, чтобы перезапустить игру. """
        # Создать списки спрайтов
        self.player_list = arcade.SpriteList()
        self.wall_list = arcade.SpriteList()  # Создать список стен

        # Настройте игрока, специально поместив его в эти координаты.
        self.player_sprite = arcade.Sprite("./Images/lama_left.png", 0.4)
        self.player_sprite.center_x = 64
        self.player_sprite.center_y = 115
        self.player_list.append(self.player_sprite)

        # Настроить стены
        self.ground_setup(32, 0, 512, ":resources:images/tiles/dirt.png")
        wall_coord = [[200, 96], [600, 160], [750, 200]]
        self.boxes_on_coords(wall_coord, ":resources:images/tiles/brickBrown.png")
        self.ground_setup(200, 814, 1102, ":resources:images/tiles/brickBrown.png")

    def on_draw(self):
        """ Метод отрисовки """
        arcade.start_render()

        # Отрисовать задний фон
        arcade.draw_lrwh_rectangle_textured(0, 0, SCREEN_WIDTH * 3, SCREEN_HEIGHT, self.background)

        # Отрисовать спрайты
        self.player_list.draw()
        self.wall_list.draw()

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажимается кнопка """

        if key == arcade.key.LEFT:
            self.player_sprite.change_x = -SPEED
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = SPEED

    def on_key_release(self, key, modifiers):
        """Вызывается, когда пользователь отпускает кнопку """

        if key == arcade.key.LEFT:
            self.player_sprite.change_x = 0
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = 0

    def update(self, delta_time):
        """ Перемещения и игровая логика """
        self.player_list.update()


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()


main()
````

# __<span style = "color: Maroon"> Физический движок </span>__

**<span style = "color: teal"> В библиотеке arcade физический движок имеет несколько функций. Нас интересуют две вещи. Во-первых, мы хотим, чтобы игрок не проходил сквозь стены. Во-вторых, мы хотим добавить в игру гравитацию. Гравитация позволит нам заставить нашего игрока прыгать и падать (если под игроком нет земли). </span>**

Мы используем физический движок, чтобы:
- избегать прохода игрока сквозь стены
- ввести в игру гравитацию
- добавить игроку возможность "прыгать"

Его обновление происходит внутри физического движка. Ниже мы показываем код, который добавляет в нашу игру физический движок </span>**

````python
import arcade

# Константы
SCREEN_WIDTH = 1100
SCREEN_HEIGHT = 700
SCREEN_TITLE = "Wall arcade"

SPEED = 5
JUMP_SPEED = 20
GRAVITY = 1


class MyGame(arcade.Window):
    """Игровой класс"""

    def __init__(self):

        # Вызвать родительский класс и настроить окно
        super().__init__(SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_TITLE)

        # Списки спрайтов
        self.player_list = None
        self.coin_list = None
        self.wall_list = None
        self.enemy_list = None

        # Отдельная переменная для спрайта игрока
        self.player_sprite = None

        # Загрузить изображение заднего фона
        self.background = arcade.load_texture("./Images/air.jpg")

    def ground_setup(self, y, x_start, x_end, pic):
        """Создать пол (много стен в одну линию)"""
        for i in range(x_start, x_end, 64):
            wall = arcade.Sprite(pic, 0.5)
            wall.center_x = i
            wall.center_y = y
            self.wall_list.append(wall)

    def boxes_on_coords(self, alist, pic):
        """Создать стены по координатам, указанным в переданном списке"""
        for i in alist:
            # Добавить квадрат на пол
            wall = arcade.Sprite(pic, 0.5)
            wall.position = i
            self.wall_list.append(wall)

    def setup(self):
        """ Настраиваем игру здесь. Вызываем эту функцию, чтобы перезапустить игру. """
        # Создать списки спрайтов
        self.player_list = arcade.SpriteList()
        self.wall_list = arcade.SpriteList()  # Создать список стен

        # Настройте игрока, специально поместив его в эти координаты.
        self.player_sprite = arcade.Sprite("./Images/lama_left.png", 0.4)
        self.player_sprite.center_x = 64
        self.player_sprite.center_y = 115
        self.player_list.append(self.player_sprite)

        # Настроить стены
        self.ground_setup(32, 0, 512, ":resources:images/tiles/dirt.png")
        wall_coord = [[200, 96], [600, 160], [750, 200]]
        self.boxes_on_coords(wall_coord, ":resources:images/tiles/brickBrown.png")
        self.ground_setup(200, 814, 1102, ":resources:images/tiles/brickBrown.png")

        # Игровой движок
        self.physics_engine = arcade.PhysicsEnginePlatformer(self.player_sprite, self.wall_list,
                                                             gravity_constant=GRAVITY)

    def on_draw(self):
        """ Метод отрисовки """
        arcade.start_render()

        # Отрисовать задний фон
        arcade.draw_lrwh_rectangle_textured(0, 0, SCREEN_WIDTH * 3, SCREEN_HEIGHT, self.background)

        # Отрисовать спрайты
        self.player_list.draw()
        self.wall_list.draw()

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажимается кнопка """

        if key == arcade.key.UP or key == arcade.key.W:
            if self.physics_engine.can_jump():
                self.player_sprite.change_y = JUMP_SPEED
        elif key == arcade.key.LEFT:
            self.player_sprite.change_x = -SPEED
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = SPEED

    def on_key_release(self, key, modifiers):
        """Вызывается, когда пользователь отпускает кнопку """

        if key == arcade.key.LEFT:
            self.player_sprite.change_x = 0
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = 0

    def update(self, delta_time):
        """ Перемещения и игровая логика """
        self.player_list.update()


def main():
    """ Основной метод """
    window = MyGame()
    window.setup()
    arcade.run()


main()
````

````python
import arcade

# Константы
SCREEN_WIDTH = 1100
SCREEN_HEIGHT = 700
SCREEN_TITLE = "Wall arcade"

SPEED = 5
JUMP_SPEED = 20
GRAVITY = 1


class Game_View(arcade.View):
    """Игровой класс."""

    def __init__(self):
        """ Инициализатор """
        # Вызвать инициализатор родительского класса
        super().__init__()

        # Списки спрайтов
        self.player_list = None
        self.coin_list = None
        self.wall_list = None
        self.enemy_list = None

        # Отдельная переменная для спрайта игрока
        self.player_sprite = None

        # Загрузить изображение заднего фона
        self.background = arcade.load_texture("./Images/air.jpg")

    def ground_setup(self, y, x_start, x_end, pic):
        """Создать пол (много стен в одну линию)"""
        for i in range(x_start, x_end, 64):
            wall = arcade.Sprite(pic, 0.5)
            wall.center_x = i
            wall.center_y = y
            self.wall_list.append(wall)

    def boxes_on_coords(self, alist, pic):
        """Создать стены по координатам, указанным в переданном списке"""
        for i in alist:
            # Добавить квадрат на пол
            wall = arcade.Sprite(pic, 0.5)
            wall.position = i
            self.wall_list.append(wall)

    def setup(self):
        """ Настраиваем игру здесь. Вызываем эту функцию, чтобы перезапустить игру. """
        # Создать списки спрайтов
        self.player_list = arcade.SpriteList()
        self.wall_list = arcade.SpriteList()  # Создать список стен

        # Настройте игрока, специально поместив его в эти координаты.
        self.player_sprite = arcade.Sprite("./Images/lama_left.png", 0.4)
        self.player_sprite.center_x = 64
        self.player_sprite.center_y = 115
        self.player_list.append(self.player_sprite)

        # Настроить стены
        self.ground_setup(32, 0, 512, ":resources:images/tiles/dirt.png")
        wall_coord = [[200, 96], [600, 160], [750, 200]]
        self.boxes_on_coords(wall_coord, ":resources:images/tiles/brickBrown.png")
        self.ground_setup(200, 814, 1102, ":resources:images/tiles/brickBrown.png")

        # Игровой движок
        self.physics_engine = arcade.PhysicsEnginePlatformer(self.player_sprite, self.wall_list,
                                                             gravity_constant=GRAVITY)

    def on_draw(self):
        """ Метод отрисовки """
        arcade.start_render()

        # Отрисовать задний фон
        arcade.draw_lrwh_rectangle_textured(0, 0, SCREEN_WIDTH * 3, SCREEN_HEIGHT, self.background)

        # Отрисовать спрайты
        self.player_list.draw()
        self.wall_list.draw()

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажимается кнопка """

        if key == arcade.key.UP or key == arcade.key.W:
            if self.physics_engine.can_jump():
                self.player_sprite.change_y = JUMP_SPEED
        elif key == arcade.key.LEFT:
            self.player_sprite.change_x = -SPEED
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = SPEED

    def on_key_release(self, key, modifiers):
        """Вызывается, когда пользователь отпускает кнопку """

        if key == arcade.key.LEFT:
            self.player_sprite.change_x = 0
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = 0

    def update(self, delta_time):
        """ Перемещения и игровая логика """
        self.player_list.update()


def main():
    """ Основной метод """
    window = arcade.Window(SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_TITLE)
    start_view = Game_View()
    window.show_view(start_view)
    start_view.setup()
    arcade.run()


main()
````

# __<span style = "color: Maroon"> Область просмотра </span>__

Область просмотра позволяет нам прокручивать окно.

**<span style = "color: teal"> Нам нужно создать пару глобальных и локальных переменных. - Глобальные переменные для прокрутки экрана
- Локальные переменные в методе _init_ для отслеживания прокрутки

````python
# Ваш код тут
import arcade

# Константы
SCREEN_WIDTH = 1100
SCREEN_HEIGHT = 700
SCREEN_TITLE = "Wall arcade"

SPEED = 5
JUMP_SPEED = 20
GRAVITY = 1

MARGIN = 200


class Intro_View(arcade.View):
    """ Представление для отображения начальноого экрана """

    def on_show(self):
        """ Вызывается один раз, когда мы переключаемся на это окно """
        arcade.set_background_color(arcade.csscolor.DARK_BLUE)

        arcade.set_viewport(0, SCREEN_WIDTH - 1, 0, SCREEN_HEIGHT - 1)

    def on_draw(self):
        """ Отрисовать это представление """
        arcade.start_render()
        arcade.draw_text("Добро пожаловать в игру", SCREEN_WIDTH / 2, SCREEN_HEIGHT / 2,
                         arcade.color.WHITE, font_size=70, anchor_x="center")
        arcade.draw_text("Чтобы начать игру, нажмите ENTER", SCREEN_WIDTH / 2, SCREEN_HEIGHT / 2 - 75,
                         arcade.color.WHITE, font_size=35, anchor_x="center")

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажимается кнопка """
        if key == arcade.key.ENTER:
            game_view = Game_View()
            game_view.setup()
            self.window.show_view(game_view)


class GameOver_View(arcade.View):
    """ Представление для отображение экрана окончания игры """

    def on_show(self):
        """ Вызывается один раз, когда мы переключаемся на это окно """
        arcade.set_background_color(arcade.csscolor.DARK_BLUE)

        arcade.set_viewport(0, SCREEN_WIDTH - 1, 0, SCREEN_HEIGHT - 1)

    def on_draw(self):
        """ Отрисовать это представление """
        arcade.start_render()
        arcade.draw_text("Игра окончена", SCREEN_WIDTH / 2, SCREEN_HEIGHT / 2,
                         arcade.color.WHITE, font_size=70, anchor_x="center")
        arcade.draw_text("Чтобы перезапустить игру, нажмите R", SCREEN_WIDTH / 2, SCREEN_HEIGHT / 2 - 75,
                         arcade.color.WHITE, font_size=35, anchor_x="center")

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажимается кнопка """
        if key == arcade.key.R:
            game_view = Game_View()
            game_view.setup()
            self.window.show_view(game_view)


class Game_View(arcade.View):
    """Игровой класс."""

    def __init__(self):
        """ Инициализатор """
        # Вызвать инициализатор родительского класса
        super().__init__()

        # Списки спрайтов
        self.player_list = None
        self.coin_list = None
        self.wall_list = None
        self.enemy_list = None

        # Отдельная переменная для спрайта игрока
        self.player_sprite = None

        # Загрузить изображение заднего фона
        self.background = arcade.load_texture("./Images/air.jpg")

    def ground_setup(self, y, x_start, x_end, pic):
        """Создать пол (много стен в одну линию)"""
        for i in range(x_start, x_end, 64):
            wall = arcade.Sprite(pic, 0.5)
            wall.center_x = i
            wall.center_y = y
            self.wall_list.append(wall)

    def boxes_on_coords(self, alist, pic):
        """Создать стены по координатам, указанным в переданном списке"""
        for i in alist:
            # Добавить квадрат на пол
            wall = arcade.Sprite(pic, 0.5)
            wall.position = i
            self.wall_list.append(wall)

    def setup(self):
        """ Настраиваем игру здесь. Вызываем эту функцию, чтобы перезапустить игру. """
        # Создать списки спрайтов
        self.player_list = arcade.SpriteList()
        self.wall_list = arcade.SpriteList()  # Создать список стен

        # Настройте игрока, специально поместив его в эти координаты.
        self.player_sprite = arcade.Sprite("./Images/lama_left.png", 0.4)
        self.player_sprite.center_x = 64
        self.player_sprite.center_y = 115
        self.player_list.append(self.player_sprite)

        # Настроить стены
        self.ground_setup(32, 0, 512, ":resources:images/tiles/dirt.png")
        wall_coord = [[200, 96], [600, 160], [750, 200]]
        self.boxes_on_coords(wall_coord, ":resources:images/tiles/brickBrown.png")
        self.ground_setup(200, 814, 1102, ":resources:images/tiles/brickBrown.png")

        # Игровой движок
        self.physics_engine = arcade.PhysicsEnginePlatformer(self.player_sprite, self.wall_list,
                                                             gravity_constant=GRAVITY)

        # Используется для контроля за прокруткой
        self.view_vert = 0
        self.view_horiz = 0

    def on_draw(self):
        """ Метод отрисовки """
        arcade.start_render()

        # Отрисовать задний фон
        arcade.draw_lrwh_rectangle_textured(-200, 0, SCREEN_WIDTH * 3, SCREEN_HEIGHT, self.background)

        # Отрисовать спрайты
        self.player_list.draw()
        self.wall_list.draw()

    def on_key_press(self, key, modifiers):
        """Вызывается всякий раз, когда нажимается кнопка """

        if key == arcade.key.UP or key == arcade.key.W:
            if self.physics_engine.can_jump():
                self.player_sprite.change_y = JUMP_SPEED
        elif key == arcade.key.LEFT:
            self.player_sprite.change_x = -SPEED
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = SPEED

    def on_key_release(self, key, modifiers):
        """Вызывается, когда пользователь отпускает кнопку """

        if key == arcade.key.LEFT:
            self.player_sprite.change_x = 0
        elif key == arcade.key.RIGHT:
            self.player_sprite.change_x = 0

    def scroll(self):
        changed = False

        # Прокрутка влево
        left_boundary = self.view_horiz + MARGIN
        if self.player_sprite.left < left_boundary:
            self.view_horiz -= left_boundary - self.player_sprite.left
            changed = True

        # Прокрутка вправо
        right_boundary = self.view_horiz + SCREEN_WIDTH - MARGIN
        if self.player_sprite.right > right_boundary:
            self.view_horiz += self.player_sprite.right - right_boundary
            changed = True

        if changed:
            # Прокрутить до целых чисел
            self.view_vert = int(self.view_vert)
            self.view_horiz = int(self.view_horiz)

            # Сделать прокрутку
            arcade.set_viewport(self.view_horiz, SCREEN_WIDTH + self.view_horiz,
                                self.view_vert, SCREEN_HEIGHT + self.view_vert)

    def update(self, delta_time):
        """ Перемещения и игровая логика """
        self.physics_engine.update()

        if self.player_sprite.center_y < 0:
            over_View = GameOver_View()
            self.window.show_view(over_View)
        self.scroll()


def main():
    """ Основной метод """
    window = arcade.Window(SCREEN_WIDTH, SCREEN_HEIGHT, SCREEN_TITLE)
    start_view = Intro_View()
    window.show_view(start_view)
    arcade.run()


main()
#
````
